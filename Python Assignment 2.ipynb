```python
# droping  and "unnamed: 8" columns
del my_data['Unnamed: 0']
del my_data['Unnamed: 8']

my_data
```

```python
# Changing weird column names

new_columns = [
    'imbd_id', 
    'title', 
    'release_year', 
    'genre', 
    'duration', 
    'country', 
    'content_rating', 
    'director', 
    'income', 
    'votes', 
    'score'
]

my_data.columns = new_columns
my_data
```

```python
# Remove empty row

my_data.dropna(how='all', inplace=True)
my_data
```

```python
# Remove symbols besides English characters, numbers and punctuation

import re
columns_to_clean = ['title', 'genre', 'director']
for col in columns_to_clean:
    my_data[col] = my_data[col].apply(lambda text: re.sub(r'[^\x00-\x7F]+', '', text))
my_data
```

```python
# Clean date column

my_data['new_date'] = pd.to_datetime(my_data['release_year'], errors='coerce')
my_data
```

```python
my_data[my_data.new_date.isna()]
```

```python
my_data.loc[70, 'release_year'] = '1951-03-06'
my_data['new_date'] = pd.to_datetime(my_data['release_year'], errors='coerce')
my_data[my_data.new_date.isna()]
```

```python
# two row have mistakes in dates, so let's remove it
my_data.dropna(subset=['new_date'], inplace=True)
my_data
```

```python
# Now lets change the date format
my_data['release_date'] = my_data['new_date'].dt.strftime('%d/%m/%Y')
del my_data['release_year']
del my_data['new_date']
my_data
```

```python
# Clean the Duration Column
import numpy as np
my_data['duration'] = my_data['duration'].str.replace(r'\D', '', regex=True)
my_data['duration'] = my_data['duration'].replace('', np.nan)
my_data['duration'] = my_data['duration'].fillna(-1).astype(int) # Filling absent durations as -1
my_data
```

```python
# Clean the Country Column

my_data['country'] = my_data['country'].str.replace(r'[^a-zA-Z ]', '', regex=True)
replacements = {'New Zeland': 'New Zealand', 'New Zesland': 'New Zealand', 'US': 'USA', 'US.': 'USA'}
my_data['country'] =my_data['country'].replace(replacements)
my_data
```

```python
# Clean the Rating Column

replacements_rating = {'Approved': '', 'Not Rated': '', 'Unrated': '', np.nan: ''}
my_data['content_rating'] = my_data['content_rating'].replace(replacements_rating)
my_data
```

```python
# Clean the Votes Column

my_data['votes'] = my_data['votes'].str.replace(r'\D', '', regex=True)
my_data['votes'] = my_data['votes'].replace('', np.nan)
my_data['votes'] = my_data['votes'].fillna(-1).astype(int)
my_data
```

```python
# Score

my_data['score'] = (my_data['score']
                 .str.replace(r'[,:]', '.', regex=True)
                 .str.lstrip('0')
                 .str.rstrip('.')
                 .str.replace(r'[^\d.]', '', regex=True)
                 .str.replace(r'\.{2,}', '.', regex=True)
                 .astype(float)
                )
my_data
```

```python
# Income

my_data['income'] = my_data['income'].replace({'o': '0'})
my_data['income'] = my_data['income'].str.replace(r'\D', '', regex=True)
my_data['income'] = my_data['income'].replace('', np.nan)
my_data['income'] = my_data['income'].fillna(-1).astype('int64')
my_data
```
